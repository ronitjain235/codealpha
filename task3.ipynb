```python
# Task 3: Data Visualization - Extended Version
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset (Titanic dataset from Kaggle or seaborn built-in)
df = sns.load_dataset("titanic")

# Show first rows
print(df.head())
print(df.info())

# -----------------------------
# 1. Basic Count of Survivors
# -----------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="survived", data=df, palette="pastel")
plt.title("Survival Count (0 = Died, 1 = Survived)")
plt.show()

# -----------------------------
# 2. Survival by Gender
# -----------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="sex", hue="survived", data=df, palette="muted")
plt.title("Survival by Gender")
plt.show()

# -----------------------------
# 3. Survival by Class
# -----------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="pclass", hue="survived", data=df, palette="cool")
plt.title("Survival by Passenger Class")
plt.show()

# -----------------------------
# 4. Age Distribution
# -----------------------------
plt.figure(figsize=(6,4))
sns.histplot(df["age"].dropna(), bins=30, kde=True, color="skyblue")
plt.title("Age Distribution of Passengers")
plt.show()

# -----------------------------
# 5. Age vs Survival
# -----------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="survived", y="age", data=df, palette="Set2")
plt.title("Age Distribution by Survival")
plt.show()

# -----------------------------
# 6. Fare Distribution
# -----------------------------
plt.figure(figsize=(6,4))
sns.histplot(df["fare"], bins=40, kde=True, color="orange")
plt.title("Fare Distribution")
plt.show()

# -----------------------------
# 7. Fare vs Class
# -----------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="pclass", y="fare", data=df, palette="Set3")
plt.title("Fare Paid by Passenger Class")
plt.show()

# -----------------------------
# 8. Survival by Embarked Port
# -----------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="embarked", hue="survived", data=df, palette="pastel")
plt.title("Survival by Boarding Port (C = Cherbourg, Q = Queenstown, S = Southampton)")
plt.show()

# -----------------------------
# 9. Correlation Heatmap
# -----------------------------
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()

# -----------------------------
# 10. Pairplot (Multiple Relations)
# -----------------------------
sns.pairplot(df[["age","fare","pclass","survived"]].dropna(), hue="survived", palette="husl")
plt.suptitle("Pairplot of Age, Fare, Class vs Survival", y=1.02)
plt.show()
```
